#0----->represents legit transactions

```python
legit=credit_card_data[credit_card_data.IsLegit==0]
fradulent=credit_card_data[credit_card_data.IsLegit==1]
```

```python
print(legit_sample.shape)
```

```python
#conactenate two dataframes
```
